```python
from langchain_community.document_loaders import GitLoader

loader=GitLoader(clone_url="https://github.com/rishim9816/Web-Development-Project.git",branch="master",repo_path="./web_project_repo/")
data=loader.load()
print(data)
```

```python


print("--- Project File Structure ---")


all_file_paths = [doc.metadata['file_path'] for doc in data]


for path in sorted(all_file_paths):
    print(path)

print("----------------------------")
```
